### **Introduction**

Coordination chemistry focuses on the study of coordination compounds/coordination complexes. These compounds consist of a central metal atom or ion bonded to surrounding molecules or ions that are called ligands.

The metalâ€“ligand bonds can be formed through coordinate covalent bonds, where the ligand donates a pair of electrons to the metal center. This branch of chemistry plays a fundamental role in fields such as bioinorganic chemistry, catalysis, material science, and medicine.

Key concepts in coordination chemistry include:
- **Coordination Number**: The number of ligand atoms directly bonded to the metal.
- **Ligand Types**: Monodentate (one donor atom), bidentate (two), and polydentate ligands.
- **Geometries**: Common spatial arrangements such as octahedral, tetrahedral, and square planar.
- **Crystal Field Theory**: A model used to describe the splitting of d-orbital energies in different ligand environments.

### **Functions**

First of all, **complexdrawerfinal.py** is a function which, by determining the path to the available metals database, can draw coordination complexes in three geometries: octahedral, tetrahedral and square planar, the complexes can be monodentate or bidentate. This function gets the ligands data from sdf files. Below, the function was imported to show what it does. It can be executed to show other complexes.

```python
from complexdrawerfinal import create_complex_from_ligand_dict
from PIL import Image
import matplotlib.pyplot as plt

# Input from user
metal = input("Enter metal (e.g., Fe): ").strip()
ligand_input = input("Enter ligands as name:count, e.g. py:4,bipy:1: ").strip()
geometry = input("Enter geometry (octahedral, tetrahedral, square planar): ").strip().lower()

# Convert ligand input into dictionary
try:
    ligand_counts = {
        name.strip(): int(count)
        for pair in ligand_input.split(',')
        if ':' in pair
        for name, count in [pair.split(':')]
    }
except Exception as e:
    print("Error parsing ligands:", e)
    ligand_counts = {}

# Create and display complex
try:
    img = create_complex_from_ligand_dict(metal, ligand_counts, geometry)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"{metal} complex with geometry: {geometry}")
    plt.show()
except Exception as e:
    print("Error drawing complex:", e)
```
