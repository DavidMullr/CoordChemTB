### **Introduction**

Coordination chemistry focuses on the study of coordination compounds/coordination complexes. These compounds consist of a central metal atom or ion bonded to surrounding molecules or ions that are called ligands.

The metal‚Äìligand bonds can be formed through coordinate covalent bonds, where the ligand donates a pair of electrons to the metal center. This branch of chemistry plays a fundamental role in fields such as bioinorganic chemistry, catalysis, material science, and medicine.

Key concepts in coordination chemistry that we are going to use include:
- **Ligand Types**: Monodentate (one donor atom), bidentate (two), and polydentate ligands (important for correct drawing of complexes).
- **Geometries**: Common spatial arrangements such as octahedral, tetrahedral, and square planar.
- **Crystal Field Theory**: A model used to describe the splitting of d-orbital energies in different ligand environments.
- **Metal center oxidation state**: Important informatiom used to calculate the total complex charg
- **...**

### **DISCUTION**

Overall, our complexorbitalssplitting function works amazing successufuly representing splitting diagrams both in low and high spin for several geometries. The only weak point of this file is that the used delta octs were estimated so might not me the most precise. Then our generalligands file works amazing as well even giving the user the option to import their own ligands. Our ligand/metal info file also works well giving the essential infos needed for coordination chemitrsy. 

Our core file also works great successfully assembling all the parts of our code together even being able to launch the GUI. The GUI also works great giving an easy to use and good looking interface for users.

The almost only and biggest flaw of our code is the complexdrawerfinal.py file. In some cases the ligands flip over towards the origin (the metal) and just overlap with the Metal center. Despite our best efforts we haven't been able to find a fix to this because it only happens with certain ligands whilst most of them are correctly orientated. Overall, rdkit is not well suited for coordination chemistry which was one of our main driving motivations to do this project. Overall, we still did a pretty good job a representing clean and professional looking complexes with monodentate and bidentate ligands in several geometric configurations. Talking about geometric configurations, we could have added more of them if we had had more time. We could also have worked on adding compatibility with tridentate and other polidentate ligands and with the macrocyclic effect. The last point that could be improved is that the function that dettects the optimal geometric configuration only looks at the metal center (and the number of ligands), it doesn't look at the properties of the ligands (strong/weak field), so a further improved could be adding configuration dependencie on the ligand properties.

The final flaw of our code is probably that it is not optimized at all, you will probably find some useless functions that were left behind after making some changes especially in the complex drawer as it was the part we had the most trouble with.
